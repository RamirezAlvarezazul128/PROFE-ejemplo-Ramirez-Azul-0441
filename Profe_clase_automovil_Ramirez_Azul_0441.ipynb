```python
class Automovil:
    def __init__(self, marca, placa, precio):
        self.marca = marca
        self.placa = placa
        self.precio = precio

    def leerdatos(self):
        self.marca = input("Ingrese la marca del automovil: ")
        self.placa = input("Ingrese la placa del automovil: ")
        self.precio = input("Ingrese el precio del automovil: ")

    def mostrardatos(self):
        print(f"Marca: {self.marca}")
        print(f"Placa: {self.placa}")
        print(f"Precio: {self.precio}")


print("Carro Numero 1")
carro1 = Automovil("", "", "")# Creando el objeto carro1
carro1.leerdatos()
carro1.mostrardatos()
print("--*-*-*-*-*-*-*-*-*")


print("Carro Numero 2")
carro2 = Automovil("", "", "")# Creando el objeto carro2
carro2.leerdatos()
carro2.mostrardatos()
print("--*-*-*-*-*-*-*-*-*")


print("Carro Numero 3")
carro3 = Automovil("", "", "")# Creando el objeto carro3
carro3.leerdatos()
carro3.mostrardatos()
print("--*-*-*-*-*-*-*-*-*")

print("software desarrollado por Ramirez Azul NL = 44")
```
